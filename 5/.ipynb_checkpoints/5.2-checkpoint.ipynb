```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_mse(f, x, y):
    return sum((f(x)-y)**2)/len(x)
```

```python
x_range = np.linspace(0, 10, num=100)
f = lambda x: 3*x + 5

x = np.array([2, 4, 8])
y = np.array([12.258, 17.24, 30.151])
plt.scatter(x,[12.258, 17.24, 30.151], c='r')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.plot(x_range,f(x_range))

print("mse ",my_mse(f,x, y))
```

```python
x_range = np.linspace(0, 10, num=100)
f = lambda x: 0.25*x**2 + 0.75*x +1.25

x = np.array([2, 4, 8])
y = np.array([3.688, 10.791, 20.705])
plt.scatter(x,y, c='r')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.plot(x_range,f(x_range))

print("mse ",my_mse(f,x, y))
```

```python
x_range = np.linspace(0, 14, num=100)
f = lambda x: 0.5*x**3 - 0.25*x**2 + 0.75*x +1.25

x = np.array([2, 4, 8, 10, 12])
y = np.array([4.872, 29.707, 246.971, 485.727, 840.658])
plt.scatter(x,y, c='r')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.plot(x_range,f(x_range))

print("mse ",my_mse(f,x, y))
```
