```python
from sympy import *
from sympy.plotting import plot
```

```python
init_printing(use_unicode=False, wrap_line=False, no_global =True)
```

```python
x = Symbol('x')
```

```python
f = x**2
f
```

```python
plot(f)
```

```python
f.subs(x, 100)
```

```python
f.subs(x, -100)
```

```python
f2 = 2*x**2+10*x-12
f2
```

```python
plot(f2)
```

```python
f2.subs(x, 100)
```

```python
f2.subs(x, -100)
```

```python
f3 = -2*x**2+10*x-12
f3
```

```python
plot(f3)
```

```python
f4 = -2*x**2+10*x+100
f4
```

```python
plot(f4)
```

```python
f5 = sqrt(x)
f5
```

```python
f5.subs(x, -1)
```

```python
plot(f5)
```

```python
plot(f5,(x, 0, 10))
```
